```python
# PROGRAM A - BASIC ndarray
# Example A1 - Comparison of Monthly Electricity Bills
import numpy as np
print("="*70)
print("EXAMPLE A1 : MONTHLY ELECTRICITY BILL COMPARISON")
print("="*70)

electricity_bill_2025 = np.array([1620, 1715, 1680, 1825, 1900, 1845, 1780, 1860])

electricity_bill_2026 = np.array([1695, 1760, 1735, 1880, 1945, 1910, 1835, 1925])

print("\nElectricity Bills - 2025")
print(electricity_bill_2025)

print("\nElectricity Bills - 2026")
print(electricity_bill_2026)

average_2025 = np.mean(electricity_bill_2025)
average_2026 = np.mean(electricity_bill_2026)

print("\n1. Average Comparison")
print(f"Average Bill (2025): ₹{average_2025:.2f}")
print(f"Average Bill (2026): ₹{average_2026:.2f}")
print(f"Higher Average Difference: ₹{abs(average_2026-average_2025):.2f}")

comparison_result = electricity_bill_2025 > electricity_bill_2026

print("\n2. Number of months where 2025 bill exceeded 2026 bill")
print(np.sum(comparison_result))

maximum_gap = np.max(electricity_bill_2026) - np.max(electricity_bill_2025)

print("\n3. Gap Between Maximum Bills")
print(f"Maximum Bill Difference = ₹{maximum_gap}")
```

```python
# PROGRAM C - ARRAY OF ONES
# Example C1 - Festival Discount System
import numpy as np
print("="*70)
print("EXAMPLE C1 : FESTIVAL DISCOUNT SYSTEM")
print("="*70)

product_names = np.array([
    "Laptop","Smartphone","Headphones","Backpack",
    "Shoes","Watch","Speaker","Keyboard"
])

original_prices = np.array([65000,32000,4500,2500,3200,6000,7500,2800])

price_multipliers = np.ones(8)

price_multipliers[[0,1]] = 0.90      # 10% discount
price_multipliers[[2,3]] = 0.85      # 15% discount
price_multipliers[[4,5]] = 0.80      # 20% discount
price_multipliers[[6,7]] = 0.95      # 5% discount

discounted_prices = original_prices * price_multipliers

full_revenue = np.sum(original_prices)
discounted_revenue = np.sum(discounted_prices)

discount_amount = original_prices - discounted_prices

print("\nProducts")
print(product_names)

print("\nOriginal Prices")
print(original_prices)

print("\nDiscounted Prices")
print(discounted_prices)

print("\n1. Revenue Comparison")
print("Revenue at Full Price : ₹", full_revenue)
print("Revenue After Discount : ₹", discounted_revenue)

print("\n2. Best Deal")
print(product_names[np.argmax(discount_amount)])
print("Discount : ₹", np.max(discount_amount))

revenue_loss = ((full_revenue - discounted_revenue)/full_revenue)*100

print("\n3. Percentage Revenue Lost")
print(f"{revenue_loss:.2f}%")
```

```python
# Example C2 - Semester Subject Weight System
import numpy as np
print("\n\n"+"="*70)
print("EXAMPLE C2 : SEMESTER SUBJECT WEIGHT SYSTEM")
print("="*70)

student_names = np.array(["Akhil","Bhanu","Charan","Deepa"])

student_marks = np.array([
    [82,76,88,91,84],
    [74,81,79,85,80],
    [90,89,92,94,91],
    [78,84,83,80,86]
])

subject_weights = np.ones(5)

subject_weights[[0,1]] = 1.5      # Core subjects
subject_weights[[2]] = 2.0        # Lab
subject_weights[[3]] = 1.0
subject_weights[[4]] = 1.2

simple_average = np.mean(student_marks, axis=1)

weighted_average = np.average(student_marks,
                              axis=1,
                              weights=subject_weights)

difference = weighted_average - simple_average

print("\nSimple Average")
print(simple_average)

print("\nWeighted Average")
print(weighted_average)

print("\n1. Difference Between Weighted and Simple Average")
print(difference)

print("\n2. Student Benefited Most")
print(student_names[np.argmax(difference)])

print("\nStudent Disadvantaged Most")
print(student_names[np.argmin(difference)])

print("\n3. Difference for Each Student")
print(difference)
```

```python
# Example D3 - Monte Carlo Simulation
import numpy as np
print("\n\n"+"="*70)
print("EXAMPLE D3 : MONTE CARLO MARKS SIMULATION")
print("="*70)

np.random.seed(123456)

marks_run1 = np.random.randint(35,101,1000)

fraction_above_75_run1 = (
    np.sum(marks_run1 > 75)
    / marks_run1.size
)

print("\nRun 1")
print("Fraction Above 75 =", fraction_above_75_run1)

np.random.seed(654321)

marks_run2 = np.random.randint(35,101,1000)

fraction_above_75_run2 = (
    np.sum(marks_run2 > 75)
    / marks_run2.size
)

difference = abs(
    fraction_above_75_run1 -
    fraction_above_75_run2
)

print("\nRun 2")
print("Fraction Above 75 =", fraction_above_75_run2)

print("\nDifference Between Runs")
print(difference)
```

```python
# PROGRAM F - IDENTITY MATRIX IN NUMPY
# Example F1 - The No-Change Proof
import numpy as np
print("="*70)
print("EXAMPLE F1 : THE NO-CHANGE PROOF")
print("="*70)

student_marks = np.array([
    [85, 78, 92],
    [76, 88, 81],
    [91, 84, 89]
])

identity_matrix = np.eye(3)

print("\nOriginal Matrix")
print(student_marks)

result_right = np.dot(student_marks, identity_matrix)
result_left = np.dot(identity_matrix, student_marks)

print("\nA × I")
print(result_right)

print("\nI × A")
print(result_left)

print("\n1. Does A × I = A ?")
print(np.array_equal(student_marks, result_right))

print("\n2. Does I × A = A ?")
print(np.array_equal(student_marks, result_left))
```

```python
# Example F3 - Pattern Discovery
import numpy as np
print("\n\n" + "="*70)
print("EXAMPLE F3 : PATTERN DISCOVERY")
print("="*70)

identity2 = np.eye(2)
identity3 = np.eye(3)
identity4 = np.eye(4)
identity5 = np.eye(5)

trace_values = np.array([
    np.trace(identity2),
    np.trace(identity3),
    np.trace(identity4),
    np.trace(identity5)
])

determinant_values = np.array([
    np.linalg.det(identity2),
    np.linalg.det(identity3),
    np.linalg.det(identity4),
    np.linalg.det(identity5)
])

print("\nTrace Values")
print(trace_values)

print("\nDeterminant Values")
print(determinant_values)

print("\nMatrix Sizes")
print(np.array([2,3,4,5]))
```
